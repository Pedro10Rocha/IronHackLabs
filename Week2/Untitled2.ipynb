```python
def total_range(i):
        if i <= 300:
            return 300 
        elif i <= 600:
            return 600
        elif i <= 900:
            return 900
        elif i <= 1200:
            return 1200
        elif i <= 1600:
            return 1600
        elif i <= 1900:
            return 1900
        elif i <= 2200:
            return 2200
        elif i <= 2600:
            return 2600
        elif i > 2600:
            return 2893
        
df['new_total_claim_amount'] = list(map(total_range,df['total_claim_amount']))
```
